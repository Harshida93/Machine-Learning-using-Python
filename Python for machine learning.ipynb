```python
import matplotlib.pyplot as plt
```

```python
import matplotlib as plot
```

```python
TitanicDf["Embarked"]=TitanicDf["Embarked"].astype('category')
```

```python
TitanicDf["Embarked_num"]=Titanic["Embarked"].cat.codes
```

```python
TitanicDf.dtypes
```

```python
TitanicDf["Embarked_num"]=TitanicDf["Embarked_num"].astype('int64')
```

```python
X="Embarked\n 1(Q) 2(S) 0(C)"
Y="Frequency"
Title="Embarked count"
plt.xlabel(X)
plt.ylabel(Y)
plt.title(Title)
plt.hist(TitanicDf.Embarked_num,bins=50,histtype='bar')
```

```python
#Converting Embarked column from categorical to numerical is helpful to see the count of each embarkment and which
#one is the highest.From the above graph we infer that category 'S' is the highest.
```
