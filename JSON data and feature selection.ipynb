```python
import json
def categorical_genres(dataset):
    arr=dataset['genres']
    genres_temp_all=[]
    feature_genres=['Action','Adventure','Fantasy','Science Fiction']
    for x in arr:
        
        loaded_json=json.loads(x)
        
   
        
        genres_all_single_movie=[]
        for i in range(len(loaded_json)):
            if(loaded_json[i]["name"] in feature_genres):
                genres_all_single_movie.append(loaded_json[i]["name"])
            elif(loaded_json[i]["name"] not in feature_genres):
                genres_all_single_movie.append("other")
                
        genres_temp_all.append(','.join(genres_all_single_movie))
        
    df_genres=pd.DataFrame({'Genres_all':genres_temp_all})
    return df_genres
```

```python
df_genres=categorical_genres(dataset)
```

```python
df_genres
```

```python
dataset=pd.concat([dataset,df_genres['Genres_all'].str.get_dummies(sep=',')],axis=1)
```

```python
dataset
```

```python
def categorical_countries(dataset):
    arr=dataset['production_countries']
    countries_temp_all=[]
    feature_countries=['United States of America','United Kingdom','Jamaica','Newzealand','Czech Republic']
    for x in arr:
        loaded_json=json.loads(x)
        countries_all_single_movie=[]
        for i in range(len(loaded_json)):
            if(loaded_json[i]["name"] in feature_countries):
                countries_all_single_movie.append(loaded_json[i]["name"])
            elif(loaded_json[i]["name"] not in feature_countries)and (loaded_json[i] != ''):
                countries_all_single_movie.append("other")
                
        countries_temp_all.append(','.join(countries_all_single_movie))
        
    df_countries=pd.DataFrame({'Countries_all':countries_temp_all})
    return df_countries
```

```python
df_countries=categorical_countries(dataset)
```

```python
dataset=pd.concat([dataset,df_countries['Countries_all'].str.get_dummies(sep=',')],axis=1)
```
