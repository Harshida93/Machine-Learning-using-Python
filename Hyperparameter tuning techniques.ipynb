```python
data=pd.read_csv(r'/Users/harshida/Downloads/Titanic_original.csv')
```

# 2

```python

y=data['Survived']                                    #target
```

```python
data.dropna(inplace=True)                             #Cleaning data
```

```python
data
```

```python
Embarked=pd.get_dummies(data['Embarked'])
```

```python
Embarked
```

```python
Sex=pd.get_dummies(data['Sex'])
```

```python
Pcl=pd.get_dummies(data['Pclass'])
```

```python
data=pd.concat([data,Embarked,Pcl,Sex],axis=1)
```

```python
data
```

```python
Clean_data=data.drop(['PassengerId','Pclass','Name','Ticket','Cabin','Sex','Embarked'],axis=1)
```

The above columns PassengerId,Pclass,name,Ticket,Cabin,Sex,Embarked do not have any effect on the accuracy of data. They will not change the prediction even if included.Name is a categorical variable of no levels and hence removed.The other columns do not impact the outcome.

```python
from sklearn.model_selection import train_test_split
```

```python
X=Clean_data.drop(['Survived'],axis=1)
y=Clean_data['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
LogReg=LogisticRegression()
```

```python
LogReg.fit(X_train,y_train)
```

```python
predict=LogReg.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predict)
```

with cost value of 10. There is a slight change in accuracy -first case-76% and in the second case it is 75%.Regularisation tries-
1.Not to give higher weights to any of the coefficients
2.Put 0 weight to features that are not required
l1 is the sum of weights whereas l2 is the sum of square of weights

```python
import sklearn.metrics as met
```

```python
y_pred=best_fit.best_estimator_.predict(X)
```

```python
met.confusion_matrix(y,y_pred)
```

```python
#met.accuracy_score(y,y_pred)
```

```python
#met.recall_score(y,y_pred)
```

```python
#met.precision_score(y,y_pred)
```

```python
tp, fp, fn, tn = met.confusion_matrix(y, y_pred).ravel()
```

```python
specificity=tn/(tn+fp)
```

```python
specificity
```

```python
Which means the model has the capability to reject the false outcomes
```

```python
sensitivity=tp/(tp+fn)
```

```python
sensitivity
```

```python
accuracy=(tp+tn)/(tp+tn+fp+fn)
```

```python
accuracy
```

```python
precision=tp/(tp+fp)
```

```python
precision
```

```python
a=2*precision*sensitivity
```

```python
b=precision+sensitivity
```

```python
F1_score=a/b
```

```python
F1_score
```

F1_score takes into account both false negatives(recall value) and false positives(precision value).
F1_score is at its best if it is close to 1. And in the case above F1_score is 0.67
which suggests that model is good enough
