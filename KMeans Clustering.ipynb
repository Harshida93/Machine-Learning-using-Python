K-means

K-means is a clustering technique which is based on distance of the data points.
A random pre-determined number of clusters is given as K. And the mean centers of each cluster is determined.
Distance of each data point from center is determined ad the closer data points are clustered together.
The same process is repeated for n number of times unitl there is no change in mean of the clusters.

```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import random
for i in range(100):
    centres=random.randrange(2,10,1)
    #print(centres)
    cluster_std=random.uniform(0.01,0.99)
    #print(cluster_std)
    X, y = make_blobs(n_samples=700, centers=centres,random_state=0, cluster_std=cluster_std)
    plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
list=[]
for i in range(2,16,1):
    k = i
    Estimator = KMeans(k)  
    Estimator.fit(X)
    y_target= Estimator.predict(X)
    plt.scatter(X[:, 0], X[:, 1], c=y_target, cmap='rainbow');
    print("Cluster #"+str(i),"Silhoutte score "
        +str(silhouette_score(X, Estimator.labels_, metric='euclidean', sample_size=None)))
    
    list.append(silhouette_score(X, Estimator.labels_, metric='euclidean', sample_size=None))
    
```
