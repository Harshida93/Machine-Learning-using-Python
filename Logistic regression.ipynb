# 3 & 4

```python
Clean_data=data.drop(['PassengerId','Pclass','Name','Ticket','Cabin','Sex','Embarked'],axis=1)
```

The above columns PassengerId,Pclass,name,Ticket,Cabin,Sex,Embarked do not have any effect on the accuracy of data. They will not change the prediction even if included.Name is a categorical variable of no levels and hence removed.The other columns do not impact the outcome.

# 6

```python
from sklearn.model_selection import train_test_split
```

```python
X=Clean_data.drop(['Survived'],axis=1)
y=Clean_data['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

# 7

```python
from sklearn.linear_model import LogisticRegression 
```

```python
LogReg=LogisticRegression()
```

```python
LogReg.fit(X_train,y_train)
```

```python
predict=LogReg.predict(X_test)
```

```python
predict
```

# 8

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predict)
```
